#Імпорти бібліотек

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import seaborn as sns
import keras
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import metrics
````

# Побудова моделі

РЕалзіуємо модель спадкуючи класс від tf.keras.model та виконуючи всі вимоги варіанту таки як алгоритм SGD з моментом та функція втрат MSE з регулярізаціює L1. Додатково також було додано навчання пакетам, тому що це дуже потужний параметр моделі навчання.

````python
class PolynomialRegression(tf.keras.Model):
    def __init__(self, degree, input_dim):
        super().__init__()

        self.degree = degree
        self.input_dim = input_dim
        self.num_coef = (degree + 1) * input_dim

        self.w = tf.Variable(tf.random.normal(shape=(self.num_coef, 1), stddev=0.1), trainable=True, dtype=tf.float32)

        self.loss = []

    def call(self, x):
        if len(x.shape) == 1:
            x = tf.expand_dims(x, axis=-1)

        input_degrees = [tf.pow(x, i) for i in range(self.degree + 1)]
        input_degrees = tf.concat(input_degrees, axis=1)

        y_pred = tf.matmul(input_degrees, self.w)
        return y_pred

def mse(y_true, y_pred):
    error = tf.square(tf.cast(tf.squeeze(y_true), dtype=tf.float32) - tf.cast(tf.squeeze(y_pred), dtype=tf.float32))
    mse_loss = tf.reduce_mean(error)
    return mse_loss

def train(model, x_train, y_train, epochs=100, batch_size=32, learning_rate=0.001, verbose=1, momentum=1, l1=0, path=None):
    optimizer = tf.keras.optimizers.SGD(learning_rate=learning_rate, momentum=momentum)


    for epoch in range(epochs):
        for batch_start in range(0, len(x_train), batch_size):
            x_batch = x_train[batch_start:batch_start+batch_size]
            y_batch = y_train[batch_start:batch_start+batch_size]

            with tf.GradientTape() as tape:
                current_loss = mse(y_batch, model(x_batch)) + l1 * tf.reduce_sum(tf.abs(model.w))

            gradients = tape.gradient(current_loss, model.trainable_weights)
            optimizer.apply_gradients(zip(gradients, model.trainable_weights))

            model.loss.append(current_loss)

        if verbose and (epoch + 1) % 10 == 0:
            print(f"Epoch {epoch + 1}/{epochs}, Loss: {current_loss.numpy()}")

            if path is not None:
                model.save_weights(path, overwrite=True)

    return None
````

# Датасет (а)

## Огляд датасету

Згенеруємо датасет відповідно до варіанту.

````python
X_data = np.linspace(-1, 1, 100)
num_coef = 5
coef = [10, 2, 30, 4, 5]
y_data = np.zeros_like(X_data)
for i in range(num_coef):
    y_data += coef[i] * np.power(X_data, i)
y_data += np.random.randn(*X_data.shape) * 1.5

split_index = int(0.8 * len(X_data))
X_train, X_test = X_data[:split_index], X_data[split_index:]
y_train, y_test = y_data[:split_index], y_data[split_index:]
````

````python
plt.scatter(X_train, y_train, color='blue', label='Навчальний набір')
plt.scatter(X_test, y_test, color='green', label='Тестовий набір')
plt.title('Навчальний та тестовий набори')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
````

Можна побачити, що наші данні дуже схожі на звичайну параболу.

## Застосування моделі

Застосуємо нашу моделі до данних та навчимо її.

````python
degree = 4
dimensions = 1

model = PolynomialRegression(degree, dimensions)
train(model, X_train, y_train, epochs=100, batch_size=64, learning_rate=0.8, verbose=1, momentum=0.05, l1=0.005, path=None)

y_pred = model.predict(X_data)

y_pred_test = model.predict(X_test)
loss = mse(y_pred_test, y_test)
print("Validation Loss:", loss.numpy())

plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color='blue', label='Навчальний набір')
plt.scatter(X_test, y_test, color='green', label='Тестовий набір')
plt.plot(X_data, y_pred, color='red', label='Передбачене значення')
plt.title('Навчальний та тестовий набори, та прогнозовані значення')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
````

Завантажемо датасет. Бостон більше не достпуний як "стандартний" датасет scikit learn через расизм (в датасеті враховується відсоток чорного населення як один з параметрів який впливає на ціни нерухомості), тому ми завантажемо не через функцію, а за посиланням.

````python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)

data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
````

````python
data_df = pd.DataFrame(data, columns=['CRIM', 'ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD',
                                      'TAX','PTRATIO','B','LSTAT'])
target_df = pd.DataFrame(target, columns=['Target'])
df = pd.concat([data_df, target_df], axis=1)
df
````

## Використання моделі

````python
model_2 = PolynomialRegression(degree=3, input_dim=X_train.shape[1])

train(model_2, X_train, y_train, epochs=100, batch_size=24, learning_rate=0.005, verbose=1, momentum=0.05, l1=0.01, path=None)

y_train_pred = model_2.predict(X_train)

print('R^2:',metrics.r2_score(y_train, y_train_pred))
print('Adjusted R^2:',1 - (1-metrics.r2_score(y_train, y_train_pred))*(len(y_train)-1)/(len(y_train)-X_train.shape[1]-1))
print('MAE:',metrics.mean_absolute_error(y_train, y_train_pred))
print('MSE:',metrics.mean_squared_error(y_train, y_train_pred))
print('RMSE:',np.sqrt(metrics.mean_squared_error(y_train, y_train_pred)))
````

Одразу я навів деякі метрики з допомогою scikitlearn. Метрики розраховані на тетсовому наборі. Можна побачити, що MSE значно менше ніж функція втрат, але варто пам'ятати що в мене ще додається регулярізація L1.
